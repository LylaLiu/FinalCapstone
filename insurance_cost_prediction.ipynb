```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#import the data
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
#select rows from the dataframe for plotting.  
x = insurance_data["age"].values
y = insurance_data["charges"].values

#plot a scatter plot
plt.scatter(x,y,color = "b")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("How age affects insurance costs")
plt.show()
```

```python
# fit the model
charges_model = LinearRegression()
age = insurance_data["age"].values.reshape(-1,1)
y = insurance_data["charges"].values

charges_model.fit(age,y)
```

```python
# plot the data and model
charges_pred = charges_model.predict(age)

plt.scatter(x,y,color = "b")
plt.plot(age,charges_pred,color = "r")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("How age affects insurance costs")
plt.show()
```
